```python
#Now, we will parse infomation from xmls

#saving all the different label values in dictionary
My_dict = dict(filepath=[],center_x=[], center_y=[], bb_width=[], bb_height=[])
for filename in path:


    info = xet.parse(filename)
    #Parse the root
    root = info.getroot()
    
    #Since the element we need to parse is object
    member_object = root.find('object')
    
    #The main thing we need is bndbox
    labels_info = member_object.find('bndbox')
    
    #Pass all the labels of bndbox to variables
    #Bydefault, it is received as string, we need to convert it to string
    xmin = int(labels_info.find('xmin').text)
    xmax = int(labels_info.find('xmax').text)
    ymin = int(labels_info.find('ymin').text)
    ymax = int(labels_info.find('ymax').text)
    
    #saving all the files in dictionary
    #Adding to respective location
    My_dict['filepath'].append(filename)

    #using tag filename
    name = root.find('filename').text
    filename = f'./images/{name}'

    # width and height through above information
    parser_size = root.find('size')
    width = int(parser_size.find('width').text)
    height = int(parser_size.find('height').text)
    
    # center_x, center_y, width , height
    #using mathematics
    center_x = (xmax + xmin)/(2*width)
    center_y = (ymax + ymin)/(2*height)

    bb_width = (xmax - xmin)/width
    bb_height = (ymax - ymin)/height

    #Adding to dictionary
    My_dict['center_x'].append(center_x)
    My_dict['center_y'].append(center_y)
    My_dict['bb_width'].append(bb_width)
    My_dict['bb_height'].append(bb_height)
    
My_dict
```

```python
#address for train images generation
training_location = './data_images/train'

for i in range(200):
    filename, x,y, w, h = My_dict['filepath'][i], My_dict['center_x'][i], My_dict['center_y'][i], My_dict['bb_width'][i], My_dict['bb_height'][i]  
    #To get only image name, first split path and then take the last element
    image_name = os.path.split(My_dict['filepath'][i])[-1]

    #now for getting only name not extension
    #Again split only using text and take first element
    txt_name = os.path.splitext(image_name)[0]
    
    Image_address = os.path.join(training_location,image_name)
    txt_address = os.path.join(training_location,txt_name+'.txt')
    
    # copy each image into the folder
    copy(My_dict['filepath'][i],Image_address)

    # generate .txt which has label info
    txt_file = f'0 {x} {y} {w} {h}'
    with open(txt_address,mode='w') as f:
        f.write(txt_file)
        
        f.close()
```
